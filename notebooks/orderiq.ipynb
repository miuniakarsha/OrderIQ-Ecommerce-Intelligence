```python
print("CLEANING DATA")
print("="*60)
#checking for duplicates

#Check Customers - train
print("Customers\n")
print(f"Total Rows(Before):{len(train_customers)}")
duplicate_count = train_customers.duplicated(subset='customer_id').sum() #Duplicated Rows Count
print(f"Duplicated Rows Count:{train_customers}")
train_customers= train_customers.drop_duplicates(subset='customer_id', keep='first')
print(f"Total Rows(After):{len(train_customers)}")

#Check Orders - train
print("Orders\n")
print(f"Total Rows(Before):{len(train_orders)}")
duplicate_count = train_orders.duplicated(subset='order_id').sum() #Duplicated Rows Count
print(f"Duplicated Rows Count:{train_orders}")
train_orders= train_orders.drop_duplicates(subset='order_id', keep='first')
print(f"Total Rows(After):{len(train_orders)}")

#Check Products - train
print("Products\n")
print(f"Total Rows(Before):{len(train_products)}")
duplicate_count = train_products.duplicated(subset='product_id').sum() #Duplicated Rows Count
print(f"Duplicated Rows Count:{train_products}")
train_products= train_products.drop_duplicates(subset='product_id', keep='first')
print(f"Total Rows(After):{len(train_products)}")

```

```python
print("HANDLING MISSING VALUES")
print("="*60)
print(f"Customer Table missing values:{train_customers.isnull().sum()}")#Customer Table
print(f"Order Table missing values:{train_orders.isnull().sum()}")#Order Table
train_orders['order_approved_at'] = train_orders['order_approved_at'].fillna(
    train_orders['order_purchase_timestamp'])
print("Orders cleaned\n")
print(f"Product Table missing values:{train_products.isnull().sum()}")#Product Table
train_products.loc[:,'product_category_name'] = train_products['product_category_name'].fillna('Unknown')
num_cols = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
train_products.loc[:,num_cols] = train_products[num_cols].fillna(train_products[num_cols].median())
print("Products cleaned\n")
```

```python
print("DATA CLEANING - CHECKING DATA TYPES")
print("="*60)
#Converting date columns into datetime
date_columns = ['order_purchase_timestamp','order_approved_at','order_delivered_timestamp']
for column in date_columns:
    train_orders[column]= pd.to_datetime(train_orders[column],errors='coerce')
    print("Converted {column} into datetime")
train_orders['order_estimated_delivery_date']=pd.to_datetime(train_orders['order_estimated_delivery_date'], errors='coerce').dt.date
print("Converted to date")
#Check numerical columns & make zero as the lowest
numerical_cols=['price','shipping_charges','payment_value']
for col in numerical_cols:
    if col in train_orderitems.columns:
        print(f"\n{col} - Min:{train_orderitems[col].min()}, Max:{train_orderitems[col].max()}")
        if(train_orderitems[col]<0).any():
            print(f"Found Negative values in{col}")
            train_orderitems[col]=train_orderitems[col].clip(lower=0)
            print("Fixed Negative Values")
```

```python
print("DATA CLEANING - VALIDATING RELATIONSHIPS")
print("="*60)
valid_orders = set(train_orders['order_id'])#Checking all orders are valid
order_items_orders = set(train_orderitems['order_id'])
missing_orders = order_items_orders - valid_orders
print(f"Order items with valid orders: {len(order_items_orders & valid_orders)}/{len(order_items_orders)}")
if missing_orders:
    print(f"Found {len(missing_orders)} order items with invalid IDs")
    train_orderitems = train_orderitems[train_orderitems['order_id'].isin(valid_orders)]
    print("Removed invalid order items")
#checking all order items have valid products
valid_products = set(train_products['product_id'])
order_items_products = set(train_products['product_id'])
missing_products = order_items_products - valid_products
print(f"Order items with valid products: {len(order_items_products & valid_products)}/{len(order_items_products)}")
if missing_products:
    print(f"Found {len(missing_products)} order items with invalid product IDs")
    train_orderitems = train_orderitems[train_orderitems['product_id'].isin(valid_products)]
    print("Removed invalid order items")
#Checking all orders have valid customers
valid_customers = set(train_customers['customer_id'])
orders_customers = set(train_orders['customer_id'])
missing_customers = orders_customers - valid_customers
print(f"Order items with valid customers: {len(orders_customers & valid_customers)}/{len(orders_customers)}")
if missing_customers:
    print(f"Found {len(missing_customers)} order items with invalid customer IDs")
    train_orders = train_orders[train_orders['customer_id'].isin(valid_customers)]
    print("Removed invalid orders")
```

```python
import pandas as pd
from sqlalchemy import create_engine 
engine = create_engine(
    "mssql+pyodbc://LAPTOP-8G4F56T7\\SQLEXPRESS/OrderIqDB?driver=ODBC+Driver+17+for+SQL+Server"
)
#Query
query = """
SELECT 
    o.order_id,
    o.order_status,
    o.order_purchase_timestamp,
    o.order_approved_at,
    o.order_delivered_timestamp,
    o.order_estimated_delivery_date,
    
    c.customer_id,
    c.customer_city,
    c.customer_state,
    
    -- AGGREGATED PRODUCT INFO
    COUNT(DISTINCT oi.product_id) AS total_products,
    -- Alternative to STRING_AGG for SQL Server
    STUFF(
        (SELECT DISTINCT ', ' + p2.product_category_name
         FROM order_items oi2 
         JOIN products p2 ON oi2.product_id = p2.product_id
         WHERE oi2.order_id = o.order_id
         FOR XML PATH('')), 1, 2, '') AS product_categories,
    SUM(oi.price) AS total_price,
    SUM(oi.shipping_charges) AS total_shipping,
    
    -- AGGREGATED PAYMENT INFO
    COUNT(DISTINCT pymt.payment_sequential) AS payment_count,
    STUFF(
        (SELECT DISTINCT ', ' + p2.payment_type
         FROM payments p2 
         WHERE p2.order_id = o.order_id
         FOR XML PATH('')), 1, 2, '') AS payment_types,
    SUM(pymt.payment_value) AS total_payment_value,
    MAX(pymt.payment_installments) AS max_installments,
    
    -- DERIVED FIELDS
    DATEDIFF(DAY, o.order_purchase_timestamp, o.order_delivered_timestamp) AS delivery_time_days,
    DATEDIFF(DAY, o.order_purchase_timestamp, o.order_estimated_delivery_date) AS estimated_delivery_days,
    CASE 
        WHEN o.order_delivered_timestamp > o.order_estimated_delivery_date THEN 1 
        WHEN o.order_delivered_timestamp IS NULL THEN NULL
        ELSE 0 
    END AS is_late

FROM orders o
    LEFT JOIN customers c ON o.customer_id = c.customer_id
    LEFT JOIN order_items oi ON o.order_id = oi.order_id
    LEFT JOIN products p ON oi.product_id = p.product_id
    LEFT JOIN payments pymt ON o.order_id = pymt.order_id

GROUP BY 
    o.order_id,
    o.order_status,
    o.order_purchase_timestamp,
    o.order_approved_at,
    o.order_delivered_timestamp,
    o.order_estimated_delivery_date,
    c.customer_id,
    c.customer_city,
    c.customer_state
"""

print("Creating Master Dataset")
master_df = pd.read_sql(query, engine)
print(f"Rows: {master_df.shape[0]}, Columns: {master_df.shape[1]}")
print(master_df.head())

print("\nColumn Names:")
print(master_df.columns)

print("\nMissing Values:")
print(master_df.isnull().sum())
```

```python
# ============================================
# FEATURE ENGINEERING - DELIVERY FEATURES
# ============================================

# Convert to datetime (already done in query, but safe to ensure)
master_df['order_purchase_timestamp'] = pd.to_datetime(master_df['order_purchase_timestamp'])
master_df['order_delivered_timestamp'] = pd.to_datetime(master_df['order_delivered_timestamp'])
master_df['order_estimated_delivery_date'] = pd.to_datetime(master_df['order_estimated_delivery_date'])

# Delivery time in days (using our existing columns)
master_df['delivery_time'] = master_df['delivery_time_days']  # Already calculated in SQL

# Estimated delivery time in days
master_df['estimated_delivery_time'] = master_df['estimated_delivery_days']  # Already calculated

# Delivery delay (positive = late, negative = early)
master_df['delivery_delay'] = master_df['delivery_time_days'] - master_df['estimated_delivery_days']

# ============================================
# FEATURE ENGINEERING - ORDER VALUE FEATURES
# ============================================

# Total order value (using our aggregated columns)
master_df['total_order_value'] = master_df['total_price'] + master_df['total_shipping']

# Average item price (if multiple products)
master_df['avg_item_price'] = master_df['total_price'] / master_df['total_products']
master_df['avg_item_price'] = master_df['avg_item_price'].fillna(0)  # Handle orders with no products

# ============================================
# FEATURE ENGINEERING - TIME-BASED FEATURES
# ============================================

master_df['order_year'] = master_df['order_purchase_timestamp'].dt.year
master_df['order_month'] = master_df['order_purchase_timestamp'].dt.month
master_df['order_day'] = master_df['order_purchase_timestamp'].dt.day
master_df['order_hour'] = master_df['order_purchase_timestamp'].dt.hour

# Additional time features
master_df['order_day_of_week'] = master_df['order_purchase_timestamp'].dt.dayofweek  # 0=Monday
master_df['order_weekend'] = master_df['order_day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
master_df['order_month_name'] = master_df['order_purchase_timestamp'].dt.month_name()

# ============================================
# FEATURE ENGINEERING - DELIVERY STATUS
# ============================================

master_df['is_delivered'] = master_df['order_status'].apply(lambda x: 1 if x == 'delivered' else 0)
master_df['is_late'] = master_df['delivery_delay'].apply(lambda x: 1 if x > 0 else 0)

# ============================================
# FEATURE ENGINEERING - CUSTOMER ORDER COUNT
# ============================================

master_df['customer_order_count'] = master_df.groupby('customer_id')['order_id'].transform('count')

# ============================================
# FEATURE ENGINEERING - PAYMENT FEATURES
# ============================================

# Average payment per installment
master_df['avg_payment_per_installment'] = master_df['total_payment_value'] / master_df['max_installments']
master_df['avg_payment_per_installment'] = master_df['avg_payment_per_installment'].fillna(master_df['total_payment_value'])

# ============================================
# FEATURE ENGINEERING - PRODUCT FEATURES
# ============================================

# Flag for multi-product orders
master_df['is_multi_product'] = (master_df['total_products'] > 1).astype(int)

# Shipping cost ratio
master_df['shipping_ratio'] = master_df['total_shipping'] / master_df['total_order_value']
master_df['shipping_ratio'] = master_df['shipping_ratio'].fillna(0)

# ============================================
# CLEAN UP - Handle Missing Values
# ============================================

# For undelivered orders, delivery-related fields should be None/0
master_df.loc[master_df['is_delivered'] == 0, 'delivery_time'] = None
master_df.loc[master_df['is_delivered'] == 0, 'delivery_delay'] = None
master_df.loc[master_df['is_delivered'] == 0, 'is_late'] = None

print("Feature engineering done!")
print(f"Final dataset shape: {master_df.shape}")
print(f"Columns: {master_df.columns.tolist()}")
```

```python
# ============================================
# CUSTOMER ANALYSIS
# ============================================
print("="*50)
print("CUSTOMER ANALYSIS")
print("="*50)

customer_analysis = master_df.groupby('customer_id').agg({
    'order_id': 'count',
    'total_payment_value': 'sum',  # Changed from 'payment_value'
    'total_order_value': 'sum',
    'customer_order_count': 'first'  # Already calculated
}).rename(columns={
    'order_id': 'total_orders',
    'total_payment_value': 'total_spent',
    'total_order_value': 'total_revenue'
})

print("\nTop 10 Customers by Total Spent:")
print(customer_analysis.nlargest(10, 'total_spent'))

# ============================================
# DELIVERY ANALYSIS
# ============================================
print("\n" + "="*50)
print("DELIVERY ANALYSIS")
print("="*50)

# Only analyze delivered orders
delivered_df = master_df[master_df['is_delivered'] == 1]

# On-time delivery rate (using our pre-calculated is_late)
on_time_rate = (delivered_df['is_late'] == 0).mean() * 100
print(f"On-time delivery rate: {on_time_rate:.2f}%")

# Average delivery time
avg_delivery = delivered_df['delivery_time'].mean()
print(f"Average delivery time: {avg_delivery:.1f} days")

# Late vs On-time comparison
late_vs_ontime = delivered_df.groupby('is_late').agg({
    'total_order_value': 'mean',
    'delivery_time': 'mean'
})
print("\nComparison (Late=1 vs On-time=0):")
print(late_vs_ontime)

# ============================================
# PRODUCT ANALYSIS
# ============================================
print("\n" + "="*50)
print("PRODUCT ANALYSIS")
print("="*50)

# Note: We need to explode product_categories for accurate analysis
# First, create a version with one row per product category
product_rows = []

for idx, row in master_df.iterrows():
    if row['product_categories'] and row['product_categories'] != '':
        categories = row['product_categories'].split(', ')
        for cat in categories:
            product_rows.append({
                'product_category_name': cat,
                'order_id': row['order_id'],
                'total_payment_value': row['total_payment_value'],
                'total_price': row['total_price']
            })

product_df = pd.DataFrame(product_rows)

if len(product_df) > 0:
    product_performance = product_df.groupby('product_category_name').agg({
        'order_id': 'count',
        'total_payment_value': 'sum'
    }).rename(columns={
        'order_id': 'total_orders',
        'total_payment_value': 'revenue'
    })
    
    print("\nTop 10 Product Categories by Revenue:")
    print(product_performance.nlargest(10, 'revenue'))
else:
    print("No product category data available")

# ============================================
# PAYMENT ANALYSIS
# ============================================
print("\n" + "="*50)
print("PAYMENT ANALYSIS")
print("="*50)

# Split payment_types into individual rows
payment_rows = []
for idx, row in master_df.iterrows():
    if row['payment_types'] and row['payment_types'] != '':
        payment_methods = row['payment_types'].split(', ')
        for method in payment_methods:
            payment_rows.append({
                'payment_type': method,
                'order_id': row['order_id'],
                'total_payment_value': row['total_payment_value']
            })

payment_df = pd.DataFrame(payment_rows)

if len(payment_df) > 0:
    print("\nPayment Type Distribution:")
    print(payment_df['payment_type'].value_counts())
    
    print("\nAverage Order Value by Payment Type:")
    print(payment_df.groupby('payment_type')['total_payment_value'].mean().sort_values(ascending=False))
else:
    print("No payment type data available")

# ============================================
# TIME TRENDS
# ============================================
print("\n" + "="*50)
print("TIME TRENDS")
print("="*50)

# Orders by month
monthly_orders = master_df.groupby('order_month')['order_id'].count()
print("\nMonthly Orders:")
print(monthly_orders)

# Orders by hour
hourly_orders = master_df.groupby('order_hour')['order_id'].count()
print("\nTop 5 Hours for Orders:")
print(hourly_orders.nlargest(5))

# Orders by day of week
weekly_orders = master_df.groupby('order_day_of_week')['order_id'].count()
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
print("\nOrders by Day of Week:")
for i, count in weekly_orders.items():
    print(f"  {day_names[i]}: {count}")

# Year-over-year (if multiple years)
yearly_orders = master_df.groupby('order_year')['order_id'].count()
print(f"\nYearly Orders:")
print(yearly_orders)

# ============================================
# FINAL SUMMARY
# ============================================
print("\n" + "="*50)
print("FINAL SUMMARY STATISTICS")
print("="*50)

print(f"Total Orders: {master_df['order_id'].nunique():,}")
print(f"Total Revenue: ${master_df['total_payment_value'].sum():,.2f}")
print(f"Average Order Value: ${master_df['total_payment_value'].mean():,.2f}")
print(f"Total Customers: {master_df['customer_id'].nunique():,}")
print(f"Average Products per Order: {master_df['total_products'].mean():.2f}")
print(f"Orders with Multiple Products: {master_df['is_multi_product'].sum():,} ({master_df['is_multi_product'].mean()*100:.1f}%)")
print(f"Total Shipping Revenue: ${master_df['total_shipping'].sum():,.2f}")
print(f"Average Shipping per Order: ${master_df['total_shipping'].mean():,.2f}")
```

```python
# First, make sure master_df has the correct data types
master_df_clean = master_df.copy()

# Convert date columns to string format that Power BI recognizes
date_columns = ['order_purchase_timestamp', 'order_approved_at', 
                'order_delivered_timestamp', 'order_estimated_delivery_date']

for col in date_columns:
    if col in master_df_clean.columns:
        master_df_clean[col] = pd.to_datetime(master_df_clean[col], errors='coerce')
        master_df_clean[col] = master_df_clean[col].dt.strftime('%Y-%m-%d %H:%M:%S')

# Replace NaN/None with appropriate values
master_df_clean = master_df_clean.fillna({
    'delivery_time_days': 0,
    'delivery_delay': 0,
    'is_late': 0,
    'product_categories': 'Unknown',
    'payment_types': 'Unknown'
})

# Ensure numeric columns are proper numbers
numeric_cols = ['total_products', 'total_price', 'total_shipping', 'payment_count',
                'total_payment_value', 'max_installments', 'delivery_time_days',
                'estimated_delivery_days', 'delivery_time', 'estimated_delivery_time',
                'delivery_delay', 'total_order_value', 'avg_item_price', 'order_year',
                'order_month', 'order_day', 'order_hour', 'order_day_of_week',
                'customer_order_count', 'avg_payment_per_installment', 'shipping_ratio']

for col in numeric_cols:
    if col in master_df_clean.columns:
        master_df_clean[col] = pd.to_numeric(master_df_clean[col], errors='coerce').fillna(0)

# Ensure boolean columns are integers
bool_cols = ['is_late', 'is_delivered', 'order_weekend', 'is_multi_product']
for col in bool_cols:
    if col in master_df_clean.columns:
        master_df_clean[col] = master_df_clean[col].astype(int)

# Remove any special characters from text columns
text_cols = ['order_id', 'order_status', 'customer_id', 'customer_city', 
             'customer_state', 'product_categories', 'payment_types', 'order_month_name']

for col in text_cols:
    if col in master_df_clean.columns:
        master_df_clean[col] = master_df_clean[col].astype(str).str.replace('[^\w\s,/-]', '', regex=True)

# Save with UTF-8 encoding (no BOM to avoid Power BI issues)
master_df_clean.to_csv('ecommerce_dashboard_clean.csv', index=False, encoding='utf-8')
print(" Saved cleaned CSV for Power BI")
```
